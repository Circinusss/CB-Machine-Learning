```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(10)
Y = (X-5)**2
print(X,Y)
```

```python
#Visualisation
```

```python
plt.style.use('seaborn')
plt.plot(X,Y)
plt.xlabel("X")
plt.ylabel("F(X)")
plt.show()
```
