```python
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
from mpl_toolkits import mplot3d
```

```python
X,y = make_regression(n_samples=1000,n_features=5,n_informative=1,noise=10,random_state=1)
print(X.shape,y.shape)
```

```python
plt.style.use('seaborn')
plt.subplot(1,2,1)
plt.scatter(X[:,0],y)
plt.subplot(1,2,2)
plt.scatter(X[:,1],y)
plt.show()
```

```python
fig = plt.figure(figsize=(10,7))
ax = plt.axes(projection='3d')
#create plot
ax.scatter3D(X[:,0],X[:,1],y,color='green')
plt.title('3D Scatter Plot')
plt.show()
```

**Linear Regression using Sk-Learn**

```python
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict([X[0],X[1]])
```

```python
y[0],y[1]
```

```python
model.score(X,y)
```
